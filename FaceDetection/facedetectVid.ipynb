```python
#program to detect target's face

#importing required libraries
import cv2
import mediapipe as mp

#capturing the video
cap=cv2.VideoCapture(0)

#face detection using mediapipe
mpface=mp.solutions.face_mesh
faces=mpface.FaceMesh(max_num_faces=2, min_tracking_confidence=0, min_detection_confidence=0.1)
mpDraw=mp.solutions.drawing_utils


while True:
    success, img=cap.read()

    result=faces.process(img)
    if result.multi_face_landmarks:
        for facelms in result.multi_face_landmarks:
            mpDraw.draw_landmarks(img, facelms, mpface.FACE_CONNECTIONS)

    cv2.imshow("camera", img)
    if cv2.waitKey(1) & 0xFF==ord('q'):
        cap.release()
        cv2.destroyAllWindows()
        break
```

```python
#program to detect target's face

#importing required libraries
import cv2
import mediapipe as mp
import matplotlib.pyplot as plt

#face detection using mediapipe
mpface=mp.solutions.face_mesh
faces=mpface.FaceMesh(max_num_faces=2, min_tracking_confidence=0, min_detection_confidence=0.1)
mpDraw=mp.solutions.drawing_utils

img=cv2.imread("download.jpg")

result=faces.process(img)
if result.multi_face_landmarks:
    for facelms in result.multi_face_landmarks:
        mpDraw.draw_landmarks(img, facelms, mpface.FACE_CONNECTIONS)

img=cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
```
